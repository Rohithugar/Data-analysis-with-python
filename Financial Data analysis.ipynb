# Correlation

```python
correlation = df['Mar Cap - Crore'].corr(df['Sales Qtr - Crore'])

print("Correlation between Market Capitalization and Quarterly Sales:", correlation)
```

As correlation between Market Capitalization and Quarterly Sales is 0.624 (Positive correlation)

```python
average_market_cap = df['Mar Cap - Crore'].mean()
average_sales_qtr = df['Sales Qtr - Crore'].mean()

print("Average Market Capitalization: {:.2f} Crore".format(average_market_cap))
print("Average Quarterly Sales: {:.2f} Crore".format(average_sales_qtr))
```

# Distribution of Market Capitalization

```python
import matplotlib.pyplot as plt
market_cap_data = df['Mar Cap - Crore']

fig, ax = plt.subplots()
n, bins, patches = plt.hist(market_cap_data, bins=10, color='skyblue', edgecolor='black')

plt.title('Distribution of Market Capitalization')
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('No. of Companies')

for i in range(len(n)):
    plt.text(bins[i] + 0.5 * (bins[i + 1] - bins[i]), n[i] + 5, str(int(n[i])), ha='center', va='bottom')
plt.show()
```
